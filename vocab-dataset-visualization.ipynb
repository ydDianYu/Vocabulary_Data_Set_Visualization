```r
library("carData")
library("RColorBrewer")
library("ggplot2")
blues <- brewer.pal(9, "Blues")
blue_range <- colorRampPalette(blues)
options(repr.plot.width = 6, repr.plot.height = 4)
ggplot(Vocab, aes(factor(education), fill = factor(vocabulary))) +
  geom_bar(position = "fill") +
  scale_fill_manual(values = blue_range(11)) +
  labs(x = "Education", y = "Proportion", fill = "Vocabulary",
       title = "Effect of Education on Vocabulary Test Score") +
  scale_y_continuous(expand = c(0, 0)) +
  theme(axis.title.x = element_text(margin=margin(10, 0, 0, 0)),
        axis.title.y = element_text(margin=margin(0, 10, 0, 0)),
        plot.title = element_text(margin=margin(10, 0, 20, 0)))
```

## Explanation
This bar plot captures the relationship between years of education received and performances on vocabulary tests, utilizing data collected in the U.S. General Social Surveys. Using different shades of blue to represent different vocabulary test scores, we observe that the proportion of darker blue areas increase as the years of education received increase, indicating that people who receive more education tend to have higher vocabulary test scores.
